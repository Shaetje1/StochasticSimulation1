```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S0 = 100
K = 100
r = 0.04
sigma = 0.25
T = 3
N = 1000
dt = T/N
R = 1000
```

```python
np.random.seed(0)
S = np.zeros((N+1, R))
S[0] = S0
for t in range(1, N+1):
    S[t] = S[t-1] * np.exp((r-0.5*sigma**2) * dt + sigma * np.sqrt(dt) * np.random.standard_normal(R))

payoff = np.maximum(np.mean(S, axis=0) - K, 0)

price = np.exp(-r*T) * np.sum(payoff) / R
print(price)
```
